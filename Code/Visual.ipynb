```python
def build():
    rect_sc = plt.Rectangle((-50, -30), height=60, width=80, fill=None)
    rect_sc_cover = plt.Rectangle((-54, 30), height=2, width=88, fill=None)
    rect_nz_1 = plt.Rectangle((-1, -1.5), height=3, width=1, fill=None)
    rect_nz_2 = plt.Rectangle((-4, -0.5), height=1, width=3, fill=None)
    rect_nz_cover_1 = plt.Rectangle((-5, -2.5), height=7, width=5, fill=None)
    rect_nz_cover_2 = plt.Rectangle((-6, 4.5), height=25.5, width=7, fill=None)
    rect_sc_branch = plt.Rectangle((30, -20), height=40, width=15, fill=None)
    white_line_1 = plt.Rectangle((29.9, -20), height=40, width=0.2, color="white")
    trig_skim = plt.Polygon(([45, 20], [15, 0.3], [15, -0.3], [45, -20]), fill=None)
    rect_sc_pc_border_1 = plt.Rectangle((45, -25), height=50, width=1.5, fill=None)
    rect_sc_pc_border_2 = plt.Rectangle((46.5, -17.5), height=35, width=2.5, fill=None)
    poly_pc = plt.Polygon(([49, 15], [64, 15], [64, 30], [89, 30], [89, 15], [104, 15], [104, -15], [89, -15], [89, -30], [64, -30], [64, -15], [49, -15]), fill=None)
    rect_pc_lv = plt.Rectangle((62, 30), height=2, width=29, fill=None)
    rect_lv_1 = plt.Rectangle((66, 32), height=5, width=2, fill=None)
    rect_lv_2 = plt.Rectangle((64.5, 37), height=3, width=5, fill=None)
    rect_pc_ic_border_1 = plt.Rectangle((104, -17.5), height=35, width=2.5, fill=None)
    rect_pc_ic_border_2 = plt.Rectangle((106.5, -17.5), height=35, width=2.5, fill=None)
    poly_ic = plt.Polygon(([109, 15], [124, 15], [124, 30], [149, 30], [149, 15], [164, 15], [164, -15], [149, -15], [149, -30], [124, -30], [124, -15], [109, -15]), fill=None)
    rect_ic_border = plt.Rectangle((122, 30), height=2, width=29, fill=None)
    rect_ic_oc_border_1 = plt.Rectangle((164, -17.5), height=35, width=2.5, fill=None)
    rect_ic_oc_border_2 = plt.Rectangle((166.5, -17.5), height=35, width=2.5, fill=None)
    rect_oc = plt.Rectangle((169, -15), height=30, width=12.5, fill=None)
    rect_oc_dc_border_1 = plt.Rectangle((181.5, -17.5), height=35, width=2.5, fill=None)
    rect_oc_dc_border_2 = plt.Rectangle((184, -17.5), height=35, width=2.5, fill=None)
    rect_ei = plt.Rectangle((166, -3), height=6, width=3, fill=None)
    poly_oc = plt.Polygon(([169, 5], [171, 5], [171, 28], [173, 28], [173, 5], [205.5, 5], [205.5, -5], [169, -5]), fill=None)
    rect_lv_3 = plt.Rectangle((169.5, 28), height=3, width=5, fill=None)
    poly_dc = plt.Polygon(([186.5, 15], [201.5, 15], [201.5, 30], [226.5, 30], [226.5, 15], [241.5, 15], [241.5, -15], [226.5, -15], [226.5, -30], [201.5, -30], [201.5, -15], [186.5, -15]), fill=None)
    rect_pi = plt.Rectangle((209, -3), height=6, width=3, fill=None)
    rect_pi_holder = plt.Rectangle((212, -4), height=8, width=4, fill=None)
    rect_pi_holder_1 = plt.Rectangle((210, 4), height=26, width=8, fill=None)
    rect_dc_border = plt.Rectangle((199.5, 30), height=2, width=29, fill=None)
    rect_qms = plt.Rectangle((201.5, 32), height=30, width=25, fill=None)
    rect_dc_cover_1 = plt.Rectangle((241.5, -17.5), height=35, width=2.5, fill=None)
    rect_dc_cover_2 = plt.Rectangle((244, -17.5), height=35, width=2.5, fill=None)
    rect_ow = plt.Rectangle((246.5, -4), height=8, width=0.75, fill=None)
    elip_lb = Ellipse((250, 0), width=0.75, height=4, fill=None)
    #rect_set = plt.Rectangle((-0.1, 0), height=0.2, width=15, fill=None)
    plt.gca().add_patch(rect_sc)
    plt.gca().add_patch(rect_sc_cover)
    plt.gca().add_patch(rect_nz_1)
    plt.gca().add_patch(rect_nz_2)
    plt.gca().add_patch(rect_nz_cover_1)
    plt.gca().add_patch(rect_nz_cover_2)
    plt.gca().add_patch(rect_sc_branch)
    plt.gca().add_patch(white_line_1)
    plt.gca().add_patch(trig_skim)
    plt.gca().add_patch(rect_sc_pc_border_1)
    plt.gca().add_patch(rect_sc_pc_border_2)
    plt.gca().add_patch(poly_pc)
    plt.gca().add_patch(rect_pc_lv)
    plt.gca().add_patch(rect_lv_1)
    plt.gca().add_patch(rect_lv_2)
    plt.gca().add_patch(rect_pc_ic_border_1)
    plt.gca().add_patch(rect_pc_ic_border_2)
    plt.gca().add_patch(poly_ic)
    plt.gca().add_patch(rect_ic_border)
    plt.gca().add_patch(rect_ic_oc_border_1)
    plt.gca().add_patch(rect_ic_oc_border_2)
    plt.gca().add_patch(rect_oc)
    plt.gca().add_patch(rect_oc_dc_border_1)
    plt.gca().add_patch(rect_oc_dc_border_2)
    plt.gca().add_patch(rect_ei)
    plt.gca().add_patch(poly_oc)
    plt.gca().add_patch(rect_lv_3)
    plt.gca().add_patch(poly_dc)
    plt.gca().add_patch(rect_pi)
    plt.gca().add_patch(rect_pi_holder)
    plt.gca().add_patch(rect_pi_holder_1)
    plt.gca().add_patch(rect_dc_border)
    plt.gca().add_patch(rect_qms)
    plt.gca().add_patch(rect_dc_cover_1)
    plt.gca().add_patch(rect_dc_cover_2)
    plt.gca().add_patch(rect_ow)
    plt.gca().add_patch(elip_lb)

    plt.axis('scaled')

    return fig, ax

def update(num, x, y):
    #line.set_data(x[:num], y[:num])
    #return line,
    plt.cla()
    fig, ax = build()
    patches = []

    if x[num] < 15 and x[num] >= 0:
        text = plt.gca().text(-50, 40, 'Droplet formed', fontsize=20)
        patches.append(text)
    elif x[num] >=15 and x[num] <= 68:
        text = plt.gca().text(-50, 40, '', fontsize=20)
        patches.append(text)
    elif x[num] > 68 and x[num] <= 170:
        text = plt.gca().text(-50, 40, 'Droplet captures dopant', fontsize=20)
        patches.append(text)
    elif x[num] > 170 and x[num] <= 190:
        text = plt.gca().text(-50, 40, 'Droplet has been ionised, also droplet size reduces due to helium particles', fontsize=20)
        patches.append(text)
    elif x[num] > 190 and x[num] <= 214:
        text = plt.gca().text(-50, 40, 'Laser evaporates droplet, ionised molecule can detected by mass spectrometer', fontsize=20)
        patches.append(text)


    if x[num] < 170:
        circ = plt.gca().add_patch(plt.Circle((x[num], y[num]), radius=2, color='b', fill=None))
        patches.append(circ)
    elif 170 < x[num] < 190:
        circ = plt.gca().add_patch(plt.Circle((x[num], y[num]), radius=2-(0.5*(x[num]-170)/20), color='b', fill=None))
        patches.append(circ)
    if x[num] > 189 and x[num] < 191:
        laser = plt.gca().add_patch(plt.Rectangle((250, -2), height=4, width=20, color='r'))
        patches.append(laser)
        poly_laser = plt.gca().add_patch(plt.Polygon(([250, -2], [250, 2], [190, 0]), color='r', fill=None))
        patches.append(poly_laser)
    #if(num % 5 == 0):
    d_positions = dopant_pos()
    for position in d_positions:
        patches.append(plt.gca().add_patch(plt.Circle((position[0], position[1]), radius=1, color='r')))
    h_positions = helium_pos()
    for position in h_positions:
        patches.append(plt.gca().add_patch(plt.Circle((position[0], position[1]), radius=0.5, color='grey')))
    if x[num] > 68 and x[num] < 168:
        d_circ = plt.gca().add_patch(plt.Circle((x[num], y[num]), radius=1, color='r'))
        patches.append(d_circ)
    elif x[num] > 168:
        d_circ = plt.gca().add_patch(plt.Circle((x[num], y[num]), radius=1, color='r', fill=None))
        patches.append(d_circ)
    return patches
```

```python
fig = plt.figure(figsize = (20,8))
ax = plt.axes()

x, y = droplet_movement()

fig, ax = build()

ani = FuncAnimation(fig, update, len(x), interval=10, fargs=[x, y], blit=True)
ani.save('animation_drawing.gif', fps=60)

#plt.show()
```
